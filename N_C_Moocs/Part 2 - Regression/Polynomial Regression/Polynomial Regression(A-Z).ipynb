#### first plot  a scatter plot of data then decide would a straight line fit the data or a curved line ,on that basis check for linear reg or poly reg

```python
# Polynomial Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
# this would make X behave as a matrix rather than a vector

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
# increasing degree increases accuracy sometimes
poly_reg = PolynomialFeatures(degree = 4)
# we have to fit and transform as well
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
#using lin_reg_2.predict() is not sufficcient as it is object of linear regression 
#ie it does not contain any polynomial terms
#Generalized form
#plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')

plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)# create a new array with its element incremented by 0.1
X_grid = X_grid.reshape((len(X_grid), 1))# 1 column nd no of rows equal to lenght of X_Grid
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
lin_reg.predict(6.5)
```

```python
# Predicting a new result with Polynomial Regression
lin_reg_2.predict(poly_reg.fit_transform(6.5))
```
